```r
x <- c(-5,-1,0,1,2,3,4,5,6)
print("Elementos de x:")
print(x)

print("Elementos de x menores que 0:")
print(x[x < 0])

print("Elementos de x menores o iguales que 0:")
print(x[x <= 0])

print("Elementos de x mayores o iguales a 3:")
print(x[x >= 3])

print("Elementos de x menores que 0 o mayores que 4:")
print(x[(x < 0) | (x > 4)])

print("Elementos de x mayores que 0 y menores que 4:")
print(x[(x > 0) & (x < 4)])

print("Elementos de x distintos de 0:")
print(x[x != 0])
```

- Arista 1: (1, 2)
- Arista 2: (1, 3)
- Arista 3: (2, 4)
- Arista 4: (2, 5)
- Arista 5: (3, 6)
- Arista 6: (4, 5)

Crea una matriz de adyacencia "matriz_adyacencia" en R para representar este grafo. En una matriz de adyacencia, las filas y columnas representan los nodos, y un valor 1 indica que existe una arista entre los nodos correspondientes, mientras que un valor 0 indica la ausencia de una arista.

- Calcula el grado del nodo 4 en el grafo. El grado de un nodo es la cantidad de aristas que inciden en él
- Encuentra todos los nodos adyacentes del nodo 2
- Cuenta la cantidad de nodos que tiene una arista hacia si mismos (self-loop)



```r
matriz_adyacencia <- matrix(0, nrow=6, ncol=6)
matriz_adyacencia[1, 2] <- 1
matriz_adyacencia[1, 3] <- 1
matriz_adyacencia[2, 4] <- 1
matriz_adyacencia[2, 5] <- 1
matriz_adyacencia[3, 6] <- 1
matriz_adyacencia[4, 5] <- 1

print(matriz_adyacencia)

# Grado del nodo 4
print("Grado del nodo 4:")
print(sum(matriz_adyacencia[, 4])+sum(matriz_adyacencia[4, -4]))

# Nodos adyacentes al 2
print("Nodos adyacentes al 2")
print((1:6)[(matriz_adyacencia + t(matriz_adyacencia))[2,] != 0])

# Nodos con aristas hacia si mismos
print("Número de self-loops")
print(sum(diag(matriz_adyacencia)))
```

`v <- c("A","I", "I", "J", "A", "Z", "X")`

```r
v <- c("A","I", "I", "J", "A", "Z", "X")
f <- factor(v, levels = LETTERS)
print(summary(f))
print("")

# Modifica una de las I por una F
f[2] <- "F"
print(summary(f))
print("")

# Agrega una nueva "A" en última posición
f[length(f) + 1] <- "A"
print(summary(f))
print("")

# Indica las letras que aparecen en el factor ordenadas por su frecuencia, empezando con la menos frecuente
s <- sort(summary(f), decreasing = TRUE)
print(s[s > 0])
```

```r
# Crear un data frame con 150 filas
num_filas <- 150
# Generar datos ficticios
set.seed(123)  # Para reproducibilidad
paises <- sample(c("EE. UU.", "Reino Unido", "Francia", "España", "Italia"), num_filas, replace = TRUE)
nombres <- paste("Película", 1:num_filas)
años <- sample(1980:2023, num_filas, replace = TRUE)
puntuaciones <- round(runif(num_filas, 1, 10), 1)
tematicas <- sample(c("Acción", "Drama", "Comedia", "Ciencia Ficción", "Animación"), num_filas, replace = TRUE)
directores <- paste("Director", 1:num_filas)
companias <- sample(c("Warner Bros.", "Universal Pictures", "Disney", "Sony Pictures", "Paramount Pictures"), num_filas, replace = TRUE)
presupuestos <- round(runif(num_filas, 1000000, 50000000), 2)
ganancias <- round(presupuestos * runif(num_filas, 0.5, 2.5), 2)
ratings <- sample(c("G", "PG", "PG-13", "R", "NC-17"), num_filas, replace = TRUE)  # Columna de rating ficticio

# Crear el data frame
peliculas <- data.frame(
    País = paises,
    Nombre = nombres,
    Año = años,
    Puntuación = puntuaciones,
    Temática = tematicas,
    Director = directores,
    Compañía = companias,
    Presupuesto = presupuestos,
    Ganancia = ganancias,
    Rating = ratings
)

peliculas %>% head(1)
peliculas %>% 
    filter(Rating %in% c("PG-13", "NC-17", "R")) %>% 
    group_by(País, Temática) %>% 
    summarise(
        totalFilms = n(), 
        totalBenefits = sum(Ganancia - Presupuesto),
        stdGrade = sd(Puntuación),
        meanGrade = mean(Puntuación)) %>%
    arrange(País, Temática, desc(meanGrade))  %>%
    group_by(País) %>% filter(meanGrade %in% meanGrade[1:2])
```

###  3.Crear una función "reducir" que dados dos vectores devuelva uno lista con dos componentes: 1) un vector con los elementos comunes sin repetir y 2) la cantidad de elementos totales eliminados al quitar los repetidos.

```r
reducir <-  function(x, y){
    comunes <- x[x%in%y]
    nEliminados <- length(union(x,y))-length(comunes)
    list(comunes, nEliminados)
}
reducir(1:5, 3:9)
```

```r
vyc <- function (x){
    x <- str_to_lower(x)
    vocales <- c("a","e","i","o","u")
    consonantes <- letters[!(letters%in%vocales)]
    (vocales_in <- vocales[str_detect(x, pattern=vocales)])
    (consonantes_in <- consonantes[str_detect(x, pattern=consonantes)])
    list(vocales_in, consonantes_in)
}
x <- "Esta es una frase"
x
vyc(x)
```

###  5.Crear una función "subpos" que dado un vector v y dos valores x e y (siendo y un parámetro opcional), devuelva una nuevo vector con los valores incluidos después de la aparición de la primera x (si x no está, empieza desde el principio) hasta la primera y (que aparezca después de la primera x), (si y no está o no se pasa por parámetro, termina hasta el final del vector). Ejemplos:

   ` subpos(1:4, 8, 3) -> 1 2 3`
    `subpos(1:4, 2, 8) -> 2 3 4`
    `subpos(1:4, 2) -> 2 3 4`


```r
subpos <- function(v, x, y = NA){
    x_indx <- c(which(v==x),1)[1]
    y_indx <- c(which(v[x_indx+1:length(v)]==y),length(v))[1]
    v[x_indx:y_indx]
}

subpos(1:4, 8, 3)
subpos(1:4, 2, 8)
subpos(1:4, 2)
```

###  1.Crear una función "creciente" que indique si los elementos de un array dado son estrictamente crecientes. No se permite ordenar el vector.

```r
creciente <- function(x){
    v <- c()
    for (i in 2:length(x)){
        v[i-1] <- x[i-1]<x[i]
    }
    all(v)
}
creciente(c(1,2,3,5,6,10))
creciente(c(1,2,3,1,6,10))
```

Otra manera más rápida:

```r
creciente <- function(x){
    all(x[1:length(x)-1]-x[2:length(x)]<0)
}
creciente(c(1,2,3,5,6,10))
creciente(c(1,2,3,1,6,10))
```
