    Se debe resolver recorriendo la cadena carácter a carácter.

    Por ejemplo: contar_letras("perro", "r") devuelve 2

De manera básica:

```python
def contar_letras(palabra, letra):
    counter = 0
    for char in palabra:
        counter += 1 if char==letra else 0
    return counter

#Test
print(
    contar_letras("Hola","w"),
    contar_letras("Hola","a"),
    contar_letras("Perruno","e"),
    contar_letras("Perruno","r"),
)
```

De manera más condensada:

```python
contar_letras = lambda palabra, letra: sum([letra==char for char in palabra])

#Test
print(
    contar_letras("Hola","w"),
    contar_letras("Hola","a"),
    contar_letras("Perruno","e"),
    contar_letras("Perruno","r"),
)
```

3. Un anagrama de una palabra pal1 es una palabra formada con las mismas letras que pal1 pero en orden distinto. Escribe una función anagrama(palabra1, palabra2) que determine si es una anagrama. Ejemplo: marta – trama

```python
def anagrama(pal1,pal2): 
    if len(pal1)!=len(pal2):
        return False
    else:
        letras1, letras2 = set(pal1), set(pal2)
        if sorted(letras1) != sorted(letras2):
            return False
        else:
            return sum([contar_letras(pal1,l1)==contar_letras(pal2,l1) for l1 in letras1])==len(letras1)

#Test
print(
    anagrama("amar", "rama"),
    anagrama("rata", "tara"),
    anagrama("return", "rama"),
    sep="\n"
)
```

    generar un ndarray con los valores entre 1 y N. En nuestro caso, pr ejemplo, 10 millones

    eliminar los valores de 2 en 2, luego de 3 en 3, y así sucesivamente. Mientras sea posible.

    Los que queden serán los números primos menores que N

    Implementa una función que realice la criba y pruébalo con N= 10 millones.

```python
def f(N):
    v = np.arange(1,N)
    i=2
    while sum(v!=-1)>i:
        v[2*i-1::i] = -1
        i+=1
    return v[v!=-1]
    
f(100)
```

1. Utiliza el conjunto de datos "Iris" para realizar una clasificación de especies de flores (setosa, versicolor, virginica) utilizando diferentes algoritmos de clasificación en scikit-learn. Por ejemplo: SVM, Random Forest y k-Nearest Neighbors. Evalúa el rendimiento de cada algoritmo utilizando métricas como la precisión.

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from tqdm.auto import tqdm
import numpy as np

iris = load_iris()
x = MinMaxScaler().fit_transform(iris.data)
y = iris.target
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2,shuffle=True, random_state=0)

models = {
    "knn":KNeighborsClassifier(),
    "svc":SVC(),
    "rfc":RandomForestClassifier()
}#Pongo los modelos base, sin modificar sus hiperparámetros, pues solo quiero probarlos

for model_key in tqdm(models.keys(), desc="Training models: "):
    cv = cross_validate(
        estimator = models[model_key],
        X=x_train,
        y=y_train,
        cv=10,
        n_jobs = -1
    )
    print(f"{model_key} scoring: {np.mean(cv["test_score"])}")

```

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.decomposition import PCA
from sklearn.svm import SVC
import numpy as np
from sklearn.metrics import accuracy_score
import seaborn as sns
import matplotlib.pyplot as plt

digits = load_digits()
x = digits.data     #No escalaré los datos en este caso
pca = PCA().fit(x)

#Analizemos el porcentaje de varianza explicada por cada componente
plt.figure(figsize=(16,2))
sns.heatmap(pca.explained_variance_ratio_.reshape(2,32), square=False)
plt.show()
```

Por lo que parecen tener un mayor impacto las 3 primeras componentes. Probemos aún así con las 5 primeras.
